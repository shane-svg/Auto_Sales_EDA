# **5. Exploratory Data Analysis**
---

## Sales Trends Over Time:

###  Month wise average sales


```python
data["order_date_dt"]=pd.to_datetime(data["ORDERDATE"])
data["Month"]=data["order_date_dt"].dt.month
data["year"]=data["order_date_dt"].dt.year
data["day"]=data["order_date_dt"].dt.day
data.groupby(["year","Month"])["SALES"].mean().reset_index()
sns.barplot(data=data.groupby(["year","Month"])["SALES"].mean().reset_index(),x="Month",y="SALES",hue="year")
plt.title("Month wise average sale")
plt.show()
```

###  Year wise average sales

```python
data.groupby(["year","Month"])["SALES"].mean().reset_index()
sns.barplot(data=data.groupby(["year","Month"])["SALES"].mean().reset_index(),x="year",y="SALES",hue="Month",palette="cool")
plt.title("year wise average sale")
plt.show()
```

###  What are the peak sales months?

```python
data["order_date_dt"]=pd.to_datetime(data["ORDERDATE"])
data["Month"]=data["order_date_dt"].dt.month
monthly_sales=data.groupby("Month")["SALES"].sum()
print(f"the month which has peak sales is {monthly_sales.idxmax()}")
print(f"with sales of {monthly_sales.max()}")
```

```python
data.groupby("Month")["SALES"].mean().reset_index()
sns.barplot(data=data.groupby("Month")["SALES"].sum().reset_index(),x="Month",y="SALES",palette="cool")
plt.title("month wise average sale")
plt.show()
```

## Product Line Analysis 

### Which product lines generate the most revenue? 

```python
product_sales=data.groupby("PRODUCTLINE")["SALES"].sum()
print(f"the product which generates the most revenue is {product_sales.idxmax()}")
print(f"with sales of {product_sales.max()}")
```

```python
plt.figure(figsize=(15,8))
data.groupby("PRODUCTLINE")["SALES"].sum()
sns.barplot(data=data.groupby("PRODUCTLINE")["SALES"].sum().reset_index(),x="PRODUCTLINE",y="SALES",palette="cool")
plt.title("Productline wise revenue")
plt.show()
```

### What is the average quantity ordered for each product line?

```python
plt.figure(figsize=(10,6))

sns.barplot(data=data,x="PRODUCTLINE",y="QUANTITYORDERED",palette="cool")
plt.title("Productline wise average order-size")
plt.show()
```

### Are there any noticeable trends in the popularity of specific product lines over time?

```python
plt.figure(figsize=(15,6))
data["order_date_dt"]=pd.to_datetime(data["ORDERDATE"])
data["year"]=data["order_date_dt"].dt.year
sns.barplot(data=data.groupby(["year","PRODUCTLINE"])["SALES"].mean().reset_index(),x="PRODUCTLINE",y="SALES",hue="year")
plt.title("Year wise trends of popularity of product lines")
plt.show()
```

## Order Details:

## Customer Insights:
